```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn import model_selection
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing

train_data = np.genfromtxt("C:/Users/Saachi/Desktop/CN/boston_train.csv", delimiter=",")
test_data = np.genfromtxt('C:/Users/Saachi/Desktop/CN/boston_test.csv', delimiter = ",")
train_data.shape
```

```python
def step_gradient(points, learn_rate, M):
    N = points.shape[0]
    num_col = points.shape[1]
    new_M = np.zeros(num_col)
    for i in range(N):
        x = points[i, 0:num_col-1]
        x = np.append(x, 1)
        y = points[i, num_col-1]
        for j in range(num_col):
            new_M[j] += (-2/N) * (y - (M * x).sum()) * x[j]
        M = M - (learn_rate * new_M)
    return M
```

```python
def gd(points, learn_rate, num_iter):
    cols = points.shape[1]
    M = np.zeros(cols)
    for i in range(num_iter):
        M = step_gradient(points, learn_rate, M)
        print(i, "Cost: ", cost(points, M))
    return M
```

```python
def cost(points, M):
    total_cost = 0
    N = points.shape[0]
    num_col = points.shape[1]
    for i in range(N):
        x = points[i, 0:num_col-1]
        x = np.append(x, 1)
        y = points[i,num_col-1]
        total_cost += (y - (M * x).sum()) ** 2
        
    total_cost = (1/N) * total_cost
    return total_cost
```

```python
def run():
    learning_rate = 0.00001
    num_iterations = 1000
    m = gd(train_data, learning_rate, num_iterations)
    return m[0:13], m[13]
```
